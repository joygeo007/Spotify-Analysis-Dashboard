```python
import requests
import json
import pandas as pd
import pycountry as pyc
```

```python
df = pd.read_csv("songs_normalize.csv")
df.head()
```

```python
artists = pd.Series(list(set(df.artist)),name="Artist")
artists
```

```python
import time
def get_country(artist):
    base_url = "http://musicbrainz.org/ws/2/artist/"
    url = f"{base_url}?query=artist:{artist}&fmt=json"
    response = requests.get(url)
    data = response.json()

    # Check if the artist was found
    if data['artists']:
        # Return the country of the first artist found
        return data['artists'][0].get('country', 'Unknown')
    else:
        return 'Unknown'

def f(x, delay=5):
    # run your code
    time.sleep(delay)
    return get_country(x)


# Get the country of each artist
countries = artists.apply(f)

```
